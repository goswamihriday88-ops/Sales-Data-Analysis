```python
import pandas as pd
import numpy as np
```

```python


data = {
    "Order_ID": [101, 102, 103, 104, 105, 106],
    "Customer": ["Amit", "Riya", "John", "Sara", "David", "Neha"],
    "Order_Date": [
        "2025-01-05",
        "2025-02-12",
        "2025-03-18",
        "2025-04-25",
        "2025-05-10",
        "2025-06-30"
    ],
    "Delivery_Date": [
        "2025-01-08",
        "2025-02-15",
        "2025-03-25",
        "2025-04-28",
        "2025-05-18",
        "2025-07-05"
    ],
    "Total": [2500, 1800, 3200, 2900, 4100, 1500],
    "Qty": [5, 3, 8, 6, 10, 2]
}

df = pd.DataFrame(data)

df
```

```python
df["Delivery_Date"] = pd.to_datetime(df["Delivery_Date"]);
df["Order_Date"] = pd.to_datetime(df["Order_Date"]);
```

```python
df.info()
```

```python
df["Week_day"] = df["Order_Date"].dt.weekday
df["Day_Name"] = df["Order_Date"].dt.day_name()
df["Month_End"] = df["Order_Date"].dt.is_month_end
df["Delivery-Time"] = (df["Delivery_Date"])-(df["Order_Date"])
df["Delivery-Days"] = ((df["Delivery_Date"])-(df["Order_Date"])).dt.days
```

```python
df
```

```python
df["Delivery_category"] = pd.cut(
    df["Delivery-Days"],
    bins=[0,3,5,10],
    labels=["Fast","Normal","Late"]
)
```

```python
df.head()
```

```python
df.set_index("Order_Date",inplace=True)
```

```python
df.resample("YE-MAR")["Total"].sum()
```

```python
df.resample("QE-MAR")["Total"].sum()
```

```python
df["Total"].expanding().mean()
```

```python
df["Running_Avereage"] = df["Total"].rolling(window=3).mean()
```

```python
df["Previous_Total"] = df["Total"].shift(1)
```

```python
df["Growth"] = df["Total"].diff()
```

```python
df["Growth%"] = (((df["Total"])-(df["Previous_Total"]))/(df["Previous_Total"])*100).round(2)
```
